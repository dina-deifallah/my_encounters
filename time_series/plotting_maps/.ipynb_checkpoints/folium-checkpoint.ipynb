# Folium
### Making nice-looking maps in Python

```python
GERMAN_STATES = ['Baden-Württemberg',
                 'Bavaria',
                 'Berlin',
                 'Brandenburg',
                 'Bremen',
                 'Hamburg',
                 'Hesse',
                 'Lower Saxony',
                 'Mecklenburg-Vorpommern',
                 'North Rhine-Westphalia',
                 'Rhineland-Palatinate',
                 'Saarland',
                 'Saxony',
                 'Saxony-Anhalt',
                 'Schleswig-Holstein',
                 'Thuringia'
                ]
```

```python
#pip install geopy
from geopy.geocoders import Nominatim
```

```python
geocoder = Nominatim(user_agent='user')
```

```python
# import requests
# SECRET_KEY = 'a2b30bd526c3b5962010ec6b3f0dd7bc'
# DARK_SKY_URL = f'https://api.darksky.net/forecast/{SECRET_KEY}/52.5170365,13.3888599'
# response = requests.get(DARK_SKY_URL)
# temp = response.json()['currently']['temperature']
# temp = round((temp - 32.0) / 1.8, 2)
# temp
```

```python
import requests
SECRET_KEY = 'a2b30bd526c3b5962010ec6b3f0dd7bc'
data = []
for i in GERMAN_STATES: 
    
    lat = geocoder.geocode(i).latitude
    lon = geocoder.geocode(i).longitude    
    DARK_SKY_URL = f'https://api.darksky.net/forecast/{SECRET_KEY}/{lat},{lon}'
    response = requests.get(DARK_SKY_URL)
    temp = response.json()['currently']['temperature']
    temp = round((temp - 32.0) / 1.8, 2)
    data.append((i, lat, lon, temp))
```

```python
data
```

```python
#pip install folium
```

```python
import folium
```

```python
germany_map = folium.Map(location=(50.7333163, 11.0747905),
             
                         zoom_start=6,
                         tiles='cartodbpositron')
```

```python
for state, lat, lon, temp in data:
    #tuple unpacking
    marker = folium.Marker((lat, lon),
                           tooltip= state,
                           icon=folium.DivIcon(html=f'<div style="font-size: 10px; font-weight: bold; display:flex; justify-content:center; align-items:center"> {temp} </div>'))
    circle_marker = folium.CircleMarker((lat, lon), radius=20)
    marker.add_to(germany_map) 
    circle_marker.add_to(germany_map)
```
