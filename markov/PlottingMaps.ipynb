```python
# accesses a web API to convert addresses to long, lat
from geopy.geocoders import Nominatim

loc = Nominatim(user_agent="mymap").geocode("Ullsteinstrasse 142, 12109 Berlin")
loc.address
```

```python
coord=(loc.latitude, loc.longitude)
coord
```

```python
spiced = folium.Marker(coord, popup='SPICED Academy',
            icon=folium.Icon(icon='star',
            color='blue'))

spiced.add_to(berlin) # <-- the map object
```

```python
import random

countries = ['Ukraine', 'Austria', 'Spain', 'Finland']    # fictional data, for the project u can use your temperature predictions
values = [random.randint(1,20) for i in range(4)]

```

```python
import pandas as pd

df = pd.DataFrame({'country': countries, 'value': values})
df
```

```python
world = folium.Map(location=[40.0, 10.0],
                    zoom_start=3,  # <-- bigger=closer
                    width=800, height=800,
                    tiles='CartoDB positron')
```

```python
world = folium.Map(location=[52.54, 13.36],
                    zoom_start=2,  # <-- bigger=closer
                    width=800, height=800,
                    tiles='CartoDB positron')

world.choropleth(  # a greek word ("area plot")
    geo_data='/Users/dina.deifallah/Documents/my_encounters/time_series/custom.geo.json', # <-- the file name
    name='choropleth',
    data=df,                           # <-- here your data goes in
    columns=['country', 'value'],      # <-- column names from the DF
    key_on='properties.name',          # <-- look into the GeoJSON file to find 'name'
    fill_color='Blues',               # <-- many names available, check docs
    fill_opacity=0.4,
    line_opacity=0.2,
    legend_name='Spicy Gradient'
)
folium.LayerControl().add_to(world)
world
```
