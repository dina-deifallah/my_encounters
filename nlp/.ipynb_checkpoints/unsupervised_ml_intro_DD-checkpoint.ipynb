## 1. What is Unsupervised Learning ? How does it compare to Supervised Learning ?

### 1.1 Definition

- Unsupervised Learning is a branch of ML where algorithms find unknown patterns in unlabeled data
- Unsupervised Learning algorithms can be grouped into Clustering, Association and Dimensionality Reduction algorithms
- has numerous use cases, including but not limited to:
     - anamoly and fraud detection
     - customer behaviour segmentation in e-commerce
     - recommender systems
     - dimensionality reduction for supervised learning or data visualization

### 1.2 Types of Unsupervised Learning

**1. Examples of Clustering algorithms:**

- K-means clustering
- Gaussian Mixture Models (GMMs)
- Density-Based Spatial Clustering of Applications with Noise (DBSCAN)
- Hierarchical Clustering
